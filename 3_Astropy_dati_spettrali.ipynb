Si sta usando i **dati spettrali** dell'acqua per costruire un grafico . I dati non sono reali , sono semplicemnete simulati. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulazione di un intervallo di lunghezze d'onda
wavelengths = np.linspace(300, 800, 500)  # Intervallo da 300 a 800 nm

# Generare una firma spettrale per un materiale (es. acqua)
intensity_water = np.exp(-((wavelengths - 600) ** 2) / (2 * (50 ** 2)))  # Firma Gaussiana

# Creare un DataFrame per visualizzare i dati
df_spectrum = pd.DataFrame({'Wavelength': wavelengths, 'Intensity': intensity_water})

# Plot della firma spettrale
plt.plot(df_spectrum['Wavelength'], df_spectrum['Intensity'], label='Acqua')
plt.xlabel('Lunghezza d\'onda (nm)')
plt.ylabel('Intensit√†')
plt.title('Firma Spettrale dell\'Acqua')
plt.legend()
plt.show()
```
