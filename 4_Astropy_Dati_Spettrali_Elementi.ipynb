```python
import pandas as pd
import matplotlib.pyplot as plt

# Dati spettrali degli elementi
data = {
    "Elemento": ["Idrogeno", "Elio", "Sodio", "Calcio", "Mercurio", "Rame", "Potassio", "Azoto"],
    "Lunghezze_d'onda_nm": [
        [121.6, 656.3, 486.1],
        [587.6, 447.1, 471.3],
        [589.0, 589.6],
        [422.7, 393.4],
        [546.1, 435.8, 404.7],
        [510.6, 578.2, 605.0],
        [766.5, 769.9],
        [337.1, 425.9]
    ]
}

# Creazione del DataFrame
df = pd.DataFrame(data)

# Preparazione dei dati per il grafico
elementi = []
lunghezze_donda = []

for index, row in df.iterrows():
    for lunghezza in row['Lunghezze_d\'onda_nm']:
        elementi.append(row['Elemento'])
        lunghezze_donda.append(lunghezza)

# Creazione del DataFrame finale
df_grafico = pd.DataFrame({
    "Elemento": elementi,
    "Lunghezza_d'onda_nm": lunghezze_donda
})
```

```python
# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.scatter(df_grafico["Lunghezza_d'onda_nm"], df_grafico["Elemento"], color='blue')

# Aggiunta di etichette e titolo
plt.title("Lunghezze d'Onda degli Elementi Spettrali")
plt.xlabel("Lunghezza d'Onda (nm)")
plt.ylabel("Elemento")
plt.grid(True)
plt.xlim(300, 800)  # Imposta limiti dell'asse x per una migliore visualizzazione

# Mostra il grafico
plt.show()
```
